Даны две функции: 
f(x) = x^3 - 50x и g(x) = -x^4 + 88x^2 - 241

Требуется:
1. Найти координаты точек пересечения
2. Построить графики функций в одной системе координат
3. Построить график функции пересечения

```python
from sympy import *
x = Symbol('x')
fx = x**3 - 50*x
gx = -1*x**4 + 88*x**2 - 241
fgx = fx - gx
```

Корни будут следующими

```python
roots = solve(fgx)
roots
```

Словарь с координатами x и y точек пересечения

```python
crosspoint = {}
i = 1
for j in roots:
    crosspoint[i] = {'x': round(N(j), 5), 'y': round(N(fx.evalf(subs={x: j}, chop=True)), 5)}
    i += 1
crosspoint
```

Координаты точек пересечения в виде кортежей

```python
print(*[(crosspoint[i]['x'], crosspoint[i]['y']) for i in crosspoint], sep=', ')
```

Вывод графиков функций f(x) и g(x) в одной системе координат 

```python
p = plot(fx, gx, x, show=False)
p[0].line_color = 'b'
p[1].line_color = 'g'
p.show()
```

График функции пересечения

```python
p = plot(fgx, x, show=False)
p[0].line_color = 'r'
p.show()
```

Все три графика в одной системе координат 

```python
p = plot(fx, gx, fgx, x, show=False)
p[0].line_color = 'b'
p[1].line_color = 'g'
p[2].line_color = 'r'
p.show()
```
